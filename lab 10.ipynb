```python
import matplotlib.pyplot as plt
```

```python
import xlrd 

book= xlrd.open_workbook('house_price.xls')

sheet= book.sheet_by_name('price')
price_list= []
bath_list=[]
htype_list=[]
lot_size_list=[]
year_list=[]
area_list=[]
for i in range (sheet.nrows):
    price,bath,htype,lot_size,year,area= sheet.row_values(i)
    if i !=0:
        price_list.append(price)
        bath_list.append(bath)
        htype_list.append(htype)
        lot_size_list.append(lot_size)
        year_list.append(year)
        area_list.append(area)
print(price_list)
print(bath_list)
print(htype_list)
print(lot_size_list)
print(year_list)
print(area_list)
```

```python
plt.plot(year_list,price_list)
plt.ylabel('year list')
plt.xlabel('price list')
plt.show()
```
